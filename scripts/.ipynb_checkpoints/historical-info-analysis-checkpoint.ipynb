```python
from ipywidgets import Layout
from ipyleaflet import Map, Heatmap, LayersControl, LayerGroup

dateRange = [['2020-10-12', '2020-10-19'],
             ['2020-10-19', '2020-10-26'],
             ['2020-10-26', '2020-11-02'],
             ['2020-11-02', '2020-11-09']]
heatmap = {}
m = {}
for i in range(0, len(devices)):
    m[i] = Map(center=(-22.961757, -43.406798), zoom=20, layout=Layout(width='100%', height='600px'))
    for j in range(0, len(dateRange)):
        try:
            heatmap[i] = Heatmap(name='bus #'+devices.iloc[i]['name']+' - '+dateRange[j][0]+' ate '+dateRange[j][1],
    #            locations=[[r['geolocation.lat'], r['geolocation.lon']] for i,r in data[i][(data[i]['payload.attr.alarm'] == 'powerCut')].iterrows()],
                locations=[[r['geolocation.lat'], r['geolocation.lon']] for i,r in data[i][(data[i]['payload.gpsSpeed'] == 0)&(data[i]['timestamp']>=dateRange[j][0])&(data[i]['timestamp']<dateRange[j][1])].iterrows()],
                radius=20
            )
            m[i].add_layer(heatmap[i]);
            #layer_group.add_layer(heatmap[i])

            # add locations where it entered the LSP (LongStop) status
            # data[0][data[0]['payload.attr.type'] == 'LSP']
            longstop = Heatmap(name='bus #'+devices.iloc[i]['name']+'- long stop - '+dateRange[j][0]+' ate '+dateRange[j][1],
                locations=[[r['geolocation.lat'], r['geolocation.lon']] for i,r in data[i][(data[i]['payload.attr.type'] == 'LSP')&(data[i]['timestamp']>=dateRange[j][0])&(data[i]['timestamp']<dateRange[j][1])].iterrows()],
                radius=20
            )
            m[i].add_layer(longstop);
            #layer_group.add_layer(longstop)
        except:
            pass
    
    control = LayersControl(position='topright')
    m[i].add_control(control)
    #m.add_layer(layer_group)


for i in range(0, len(devices)):
    m[i]
```
